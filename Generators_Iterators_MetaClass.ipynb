```python
def even():
  n=0
  while True:
    yield n
    n=n+2

for item in even():
  print(item)
```

```python
#creating pipelines using Generator
```

```python
def fibonacci_numbers(nums):
  x,y=0,1
  for i in range(nums):
    x,y=y,x+y
    yield x

def square(nums):
  for num in nums:
    yield num**2

print(sum(square(fibonacci_numbers(10)))) #pipeline
```

```python
fibonacci_numbers(10)
```
