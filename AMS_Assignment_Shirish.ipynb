```python
U, S, Vt = np.linalg.svd(X_red, full_matrices=False)

# Determine the number of singular values to keep
k = 1

# Create the low-rank approximation
U_k = U[:, :k]
S_k = np.diag(S[:k])
Vt_k = Vt[:k, :]

X_low_rank = np.dot(U_k, np.dot(S_k, Vt_k))

print("Original Matrix Shape:", X_red.shape)
print("Low-Rank Approximation Shape:", X_low_rank.shape)
print("Low-Rank Approximation:\n", X_low_rank)
```
